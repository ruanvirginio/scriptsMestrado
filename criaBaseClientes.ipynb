```python
import pandas as pd
import numpy as np

cols = ['UC', 'LIGACAO', 'DATA_ENCERRAMENTO_DO_CONTRATO', 'DSC_GRUPO_FORNECIMENTO', 'CLASSE']
base_cadastro = pd.read_csv(r'I:\CPOP\06.Exportacoes_SQL\02.Extracoes\A-003 - RELACAO CADASTRAL.csv', sep=';', usecols=cols)
base_cadastro
```

```python
uc_trafo = pd.read_csv(r'C:\Users\rcvs3\OneDrive - Energisa\Documentos\Carregamento de Transformadores\newV\UC_Trafo.txt', sep=';', encoding='latin-1',dtype=object)
trafo_alim = pd.read_csv(r'C:\Users\rcvs3\OneDrive - Energisa\Documentos\Carregamento de Transformadores\newV\Trafo_ID.txt', sep=';', encoding='latin-1',dtype=object)

uc_alim = pd.merge(uc_trafo, trafo_alim, on='Instalacao')
uc_alim = uc_alim[['Conta', 'Circuit_Id']]
uc_alim['Conta'] = uc_alim['Conta'].str.slice(2).astype(int)
uc_alim

uc_alim.rename(columns={
    'Conta': 'UC',
    'Circuit_Id': 'ALIMENTADOR'
}, inplace=True)

df = pd.merge(base_cadastro, uc_alim, on='UC')

#############################################################

sub = pd.read_excel(r'C:\Users\rcvs3\OneDrive - Energisa\Documentos\Carregamento de Transformadores\newV\subestacoes.xlsx')
sub['TRAFO'] = sub['SUBESTACAO'] + "_DJ_" + sub['DISJUNTOR']


df2 = pd.merge(df, sub, on='ALIMENTADOR')

# df2 = df2[['UC', 'PotenciaInstalada_MW', 'DataInicioInjecao', 'TRAFO']]
# df2['DataInicioInjecao'] = pd.to_datetime(df2['DataInicioInjecao'], format='%d/%m/%Y %H:%M:%S')

df2
```

```python
df2 = df2[['UC',	'DSC_GRUPO_FORNECIMENTO',	'CLASSE',	'LIGACAO',	'DATA_ENCERRAMENTO_DO_CONTRATO', 'TRAFO']]
df2
```

```python
import numpy as np

# Criar intervalo de datas
date_range = pd.date_range(start="2018-01-01", end="2024-12-31", freq='D')

# Garantir que as colunas de data estão no formato datetime
df2["LIGACAO"] = pd.to_datetime(df2["LIGACAO"], errors="coerce")
df2["DATA_ENCERRAMENTO_DO_CONTRATO"] = pd.to_datetime(df2["DATA_ENCERRAMENTO_DO_CONTRATO"], errors="coerce")

# Recriar o processo de contagem diária de clientes ativos
result = []

for trafo in df2["TRAFO"].unique():
    df_trafo = df2[df2["TRAFO"] == trafo]

    for date in date_range:
        active_clients = df_trafo[(df_trafo["LIGACAO"] <= date) & 
                                  ((df_trafo["DATA_ENCERRAMENTO_DO_CONTRATO"].isna()) | 
                                   (df_trafo["DATA_ENCERRAMENTO_DO_CONTRATO"] > date))]

        grouped = active_clients.groupby(["CLASSE", "DSC_GRUPO_FORNECIMENTO"]).size().reset_index(name="QTD_CLIENTES")
        grouped["DATA"] = date
        grouped["TRAFO"] = trafo
        result.append(grouped)

df_result = pd.concat(result, ignore_index=True)
df_result.head()
```

```python
df_result.to_csv('Base_Quantidade_Clientes.csv', sep=';', index=False)
```
